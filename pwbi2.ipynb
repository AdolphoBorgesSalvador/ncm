```python
cliente2 = cliente[['Cliente', 'Conceito de Pesquisa', 'CONCEITO DE PESQUISA_2']]
cliente2
```

```python
# Converte a coluna para datetime, se ainda não tiver sido convertida
zbm51_historico['Data de lançamento'] = pd.to_datetime(zbm51_historico['Data de lançamento'], errors='coerce')

# Ajusta todas as datas parlançamentoa o primeiro dia do mês
zbm51_historico['Data de lançamento'] = zbm51_historico['Data de lançamento'].apply(lambda x: x.replace(day=1))

# Formata as datas para '01/mm/aaaa'
zbm51_historico['Data de lançamento'] = zbm51_historico['Data de lançamento'].dt.strftime('01/%m/%Y')
```
